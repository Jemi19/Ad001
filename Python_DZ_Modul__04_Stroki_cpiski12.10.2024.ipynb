```python
import random

# Генерируем два списка случайных целых чисел
list1 = [random.randint(1, 100) for _ in range(10)]
list2 = [random.randint(1, 100) for _ in range(10)]

print("Список 1:", list1)
print("Список 2:", list2)

# 1. Формируем третий список, содержащий элементы обоих списков
list3 = list1 + list2
print("Список 3 (все элементы):", list3)

# Комментарий: мы используем оператор + для конкатенации двух списков

# 2. Формируем третий список, содержащий элементы обоих списков без повторений
list4 = list(set(list1 + list2))
print("Список 4 (все элементы без повторений):", list4)

# Комментарий: мы преобразуем список в множество (set), которое удаляет повторы, а затем обратно в список

# 3. Формируем третий список, содержащий общие элементы обоих списков
list5 = [element for element in list1 if element in list2]
print("Список 5 (общие элементы):", list5)

# Комментарий: мы используем генератор списка для перебора элементов первого списка и проверки, есть ли они во втором списке

# 4. Формируем третий список, содержащий уникальные элементы каждого списка
list6 = list(set(list1) ^ set(list2))
print("Список 6 (уникальные элементы):", list6)

# Комментарий: мы используем оператор симметрической разности ^ для нахождения уникальных элементов в обоих списках

# 5. Формируем третий список, содержащий минимальное и максимальное значение каждого списка
list7 = [min(list1), max(list1), min(list2), max(list2)]
print("Список 7 (мин и макс значения):", list7)

# Комментарий: мы используем функции min и max для нахождения минимального и максимального значения в каждом списке
```
