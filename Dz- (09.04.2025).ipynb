```python
def draw_triangle_left(size=5):
    """Рисует синий треугольник слева"""
    color_code = "\033[44m"
    reset_color = "\033[0m"
    for i in range(size):
        stars = '*' * (i+1)
        print(f"{color_code}{stars.ljust(20)}{reset_color}")

def draw_triangle_right(size=5):
    """Рисует синий треугольник справа"""
    color_code = "\033[44m"
    reset_color = "\033[0m"
    for i in range(size):
        stars = '*' * (i+1)
        print(f"{color_code}{stars.rjust(20)}{reset_color}")

def draw_horizontal_hourglass(size=5):
    """Рисует горизонтально ориентированные песочные часы"""
    color_code = "\033[44m"
    reset_color = "\033[0m"
    half_size = size // 2
    total_width = size * 2
    # Левый восходящий треугольник
    for i in range(half_size):
        stars = '*' * (half_size - i)
        left_padding = ' ' * i
        right_padding = ' ' * (total_width - len(left_padding) - len(stars))
        line = f"{left_padding}{stars}{right_padding}"
        print(f"{color_code}{line.center(total_width)}{reset_color}")
    # Правый нисходящий треугольник
    for i in range(half_size):
        stars = '*' * (i+1)
        left_padding = ' ' * (half_size - i - 1)
        right_padding = ' ' * (total_width - len(left_padding) - len(stars))
        line = f"{left_padding}{stars}{right_padding}"
        print(f"{color_code}{line.center(total_width)}{reset_color}")

def draw_triangle_upper_left(size=5):
    """Рисует синий треугольник в верхнем левом углу"""
    color_code = "\033[44m"
    reset_color = "\033[0m"
    for i in range(size):
        stars = '*' * (i+1)
        print(f"{color_code}{stars.ljust(20)}{reset_color}")

def draw_triangle_lower_right(size=5):
    """Рисует синий треугольник в нижнем правом углу"""
    color_code = "\033[44m"
    reset_color = "\033[0m"
    for i in reversed(range(size)):
        stars = '*' * (i+1)
        print(f"{color_code}{stars.rjust(20)}{reset_color}")

def main():
    while True:
        print("\nВыберите фигуру:")
        print("1. Треугольник вверху справа")
        print("2. Треугольник внизу слева")
        print("3. Треугольник посередине сверху")
        print("4. Треугольник посередине снизу")
        print("5. Песочные часы вертикально")
        print("6. Горизонтальные песочные часы")
        print("7. Треугольник слева")
        print("8. Треугольник справа")
        print("9. Треугольник верхний левый угол")
        print("10. Треугольник нижний правый угол")
        print("11. Выход")
        
        choice = input("Ваш выбор: ")
        
        if choice == '1':
            draw_triangle_up_right()
        elif choice == '2':
            draw_triangle_down_left()
        elif choice == '3':
            draw_triangle_center_top()
        elif choice == '4':
            draw_triangle_center_bottom()
        elif choice == '5':
            draw_hourglass()
        elif choice == '6':
            draw_horizontal_hourglass()
        elif choice == '7':
            draw_triangle_left()
        elif choice == '8':
            draw_triangle_right()
        elif choice == '9':
            draw_triangle_upper_left()
        elif choice == '10':
            draw_triangle_lower_right()
        elif choice == '11':
            break
        else:
            print("Неверный выбор!")

if __name__ == "__main__":
    main()
```
